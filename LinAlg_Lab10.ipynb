```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_quiv(x,t_mat=np.eye(2)):
    z = np.arange(0,10,0.50)
    x_prime = x @ t_mat
    size= (5,5)
    plt.figure(figsize=(5,5))

    plt.xlim(-size[0],size[0])
    plt.ylim(-size[1],size[1])
    plt.xticks(np.arange((-size[0]), size[0]+1, 1.0))
    plt.yticks(np.arange((-size[1]), size[1]+1, 1.0))

    
    plt.scatter(z*x_prime[0,0],z*x_prime[0,1],color = ['purple'])
    plt.scatter(z*x_prime[1,0],z*x_prime[1,1],color = ['blue'])
    
    
    plt.grid()
    plt.show()
```

```python
A = np.array([
    [1, 0],
    [0, 1]
])

plot_quiv(A)
```

### Repositioning/Translation

```python
t_mat = np.array([
    [-1,0],
    [0,-1]
])
plot_quiv(A)
plot_quiv(A, t_mat)
```

### Shears

```python
A = np.eye(2)
## Transformation Matrices
shear1 = np.array([
    [1,0],
    [1,1]
])
shear2 = np.array([
    [1,0],
    [1,-1]
])
plot_quiv(A)
print(A)
plot_quiv(A, shear1)
plot_quiv(A, shear2)
```

### Rotation

```python
def rot_matrix(theta):
    theta = np.deg2rad(theta)
    rot_mat = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    return rot_mat
```

```python
rotate = rot_matrix(45)
plot_quiv(A)
scale = 2*np.eye(2)
plot_quiv(scale@rotate@A)
```
